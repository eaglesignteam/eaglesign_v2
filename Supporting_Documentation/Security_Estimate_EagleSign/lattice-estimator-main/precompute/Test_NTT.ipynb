```python
def product(a,b,q,n):
    assert len(a)==len(b)==n, "Entries lengths must be equal"
    tmp = np.zeros(2*n, dtype=int)
    for i in range(n):
        for j in range(n):
            tmp[i + j] += (a[i]*b[j])
            
    for i in range(n):
        #tp = ((tmp[i]-tmp[n+i])%q)
        #tmp[i] = tp if (-q/2 < tp < q/2) else (tp - q)
        tmp[i] = ((tmp[i]+tmp[n+i])%q)
        
    return list(tmp[:n])
```

```python
def product_bitwize(a,b,q,n):
    assert len(a)==len(b)==n, "Entries lengths must be equal"
    tmp = np.zeros(n, dtype=int)
    for i in range(n):
        for j in range(n):
            #tp = ((a[i]*b[i])%q)
            #tmp[i] = tp if (-q/2 < tp < q/2) else (tp - q)
            tmp[i] = ((a[i]*b[i])%q)
        
    return list(tmp)
```
