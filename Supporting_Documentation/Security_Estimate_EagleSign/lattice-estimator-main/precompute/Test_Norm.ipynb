#### Q = 17

> a$\times$b

```python
def product(a,b,q,n):
    #assert len(a)==len(b)==n, "Entries lengths must be equal"
    tmp = np.zeros(2*n, dtype=int)
    for i in range(n):
        for j in range(n):
            tmp[i + j] += (a[i]*b[j])
            
    for i in range(n):
        tp = ((tmp[i]-tmp[n+i])%q)
        tmp[i] = tp if (-q/2 < tp < q/2) else (tp - q)
        
    return list(tmp[:n])
```

```python
def add(a,b,q,n):
    #assert len(a)==len(b)==n, "Entries lengths must be equal"
    tmp = np.zeros(n, dtype=int)            
    for i in range(n):
        tp = ((a[i]+b[i])%q)
        tmp[i] = tp if (-q/2 < tp < q/2) else (tp - q)
        
    return list(tmp[:n])
```

```python
def norm_sup(a):
    return max([abs(x) for x in a])
```

```python
def norm_1(a):
    return sum([abs(x) for x in a])
```

```python
a = [-2, 3, 0, 0, 0, -1, -3, 0]
b = [0, -1, -2, 1, 0, 0, 0, -2]
c = product(a, b, 17, 8)
```

```python
norm_sup(c)<norm_sup(a)*norm_1(b)
```

```python
norm_sup(c)<norm_sup(b)*norm_1(a)
```

```python
def random_test(n,q,N):
    for i in range(n):
        a = np.random.choice(range(-(q-1)//2, (q+1)//2), N)
        b = np.random.choice(range(-(q-1)//2, (q+1)//2), N)
        c = product(a,b,q,N)
        assert (norm_sup(c) <= min([norm_sup(a)*norm_1(b), norm_sup(b)*norm_1(a)])), f"Non vérifié pour \na = {a} \net \nb={b}"
```

```python
random_test(1000, 17, 8)
```
